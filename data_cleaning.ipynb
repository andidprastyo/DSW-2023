### Encoding

```python
columns_encoding = ['Games Product', 'Music Product', 'Education Product', 'Video Product', 'Use MyApp', 'Call Center', 'Churn Label']

le = LabelEncoder()

for column in columns_encoding:
    df[column] = le.fit_transform(df[column])
    
df.head()
```
